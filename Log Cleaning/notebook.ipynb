```python
import re
format_pat = re.compile(
    r'^'
    r'(?P<host>\d{1,3}(?:\.\d{1,3}){3})\s+'
    r'(?P<identity>\S+)\s+'
    r'(?P<user>\S+)\s+'
    r'\[(?P<time>[^\]]+)\]\s+'
    r'"(?P<request>[^"]+)"\s+'
    r'(?P<status>\d{3})\s+'
    r'(?P<bytes>-|\d+)\s+'
    r'"(?P<referer>[^"]*)"\s+'
    r'"(?P<user_agent>[^"]*)"\s+'
    r'"(?P<extra>[^"]*)"'
    r'$'
)

# From https://www.kaggle.com/datasets/eliasdabbas/web-server-access-logs?resource=download
```

```python
logPath = 'access.log'
```

```python
# Count URL access frequency
URL_counts = {}
with open(logPath, "r") as f:
    for line in (l.rstrip() for l in f):
        match = format_pat.match(line)
        if match:
             access = match.groupdict()
             request = access['request']
             if(len(request.split()) == 3):
                (action, URL, protocal) = request.split()
                if URL in URL_counts:
                    URL_counts[URL] += 1
                else:
                    URL_counts[URL] = 1
results = sorted(URL_counts, key=lambda i:int(URL_counts[i]), reverse=True)
for result in results[:20]:
    print(result + ": " + str(URL_counts[result]))
```

```python
# Count URL access frequency for GET requests only
URL_counts = {}
with open(logPath, "r") as f:
    for line in (l.rstrip() for l in f):
        match = format_pat.match(line)
        if match:
             access = match.groupdict()
             request = access['request']
             if(len(request.split()) == 3):
                (action, URL, protocal) = request.split()
                if action =="GET":
                    if URL in URL_counts:
                        URL_counts[URL] += 1
                    else:
                        URL_counts[URL] = 1
results = sorted(URL_counts, key=lambda i:int(URL_counts[i]), reverse=True)
for result in results[:20]:
    print(result + ": " + str(URL_counts[result]))
```

```python
# Count User-Agent frequency and check for bots
UserAgents = {}
with open(logPath, "r") as f:
    for line in (l.strip() for l in f):
        match = format_pat.match(line)
        if match:
            access = match.groupdict()
            agent = access['user_agent']
            if agent in UserAgents:
                UserAgents[agent] += 1
            else:
                UserAgents[agent] = 1
results = sorted(UserAgents, key=lambda i:int(UserAgents[i]), reverse=True)
for result in results[:20]:
    print(result + ": " + str(UserAgents[result]))
```

```python
# Count URL access frequency excluding bots
URL_counts = {}
with open(logPath, "r") as f:
    for line in (l.rstrip() for l in f):
        match = format_pat.match(line)
        if match:
             access = match.groupdict()
             agent = access['user_agent']
             if(not("bot" in agent.lower() or "spider" in agent.lower() or "crawl" in agent.lower() or agent == "-")):
                request = access['request']
                if(len(request.split()) == 3):
                    (action, URL, protocal) = request.split()
                    if action =="GET":
                        if URL in URL_counts:
                            URL_counts[URL] += 1
                        else:
                            URL_counts[URL] = 1
results = sorted(URL_counts, key=lambda i:int(URL_counts[i]), reverse=True)
for result in results[:20]:
    print(result + ": " + str(URL_counts[result]))
```

```python
# Now cleaning up images, css, js accesses as well
URL_counts = {}
STATIC_PREFIXES = (
    "/static/",
    "/image/",
    "/favicon.ico"
)

STATIC_EXTS = (
    ".png", ".jpg", ".jpeg", ".gif",
    ".css", ".js", ".woff", ".ico","logo"
)
with open(logPath, "r") as f:
    for line in (l.rstrip() for l in f):
        match = format_pat.match(line)
        if match:
             access = match.groupdict()
             agent = access['user_agent']
             if(not("bot" in agent.lower() or "spider" in agent.lower() or "crawl" in agent.lower() or agent == "-")):
                request = access['request']
                if(len(request.split()) == 3):
                    (action, URL, protocal) = request.split()
                    if not (URL.startswith(STATIC_PREFIXES) or URL.endswith(STATIC_EXTS)):
                        if action =="GET":
                            if URL in URL_counts:
                                URL_counts[URL] += 1
                            else:
                                URL_counts[URL] = 1
results = sorted(URL_counts, key=lambda i:int(URL_counts[i]), reverse=True)
for result in results[:50]:
    print(result + ": " + str(URL_counts[result]))
```

```python
clean_log_path = "cleaned_access.log"

with open(logPath, "r") as f, open(clean_log_path, "w", encoding="utf-8") as out:
    for line in f:
        line = line.rstrip()
        match = format_pat.match(line)
        if not match:
            continue

        access = match.groupdict()
        agent = access["user_agent"].lower()

        if "bot" in agent or "spider" in agent or "crawl" in agent:
            continue

        request = access["request"]
        parts = request.split()
        if len(parts) != 3:
            continue

        action, URL, protocol = parts
        if action != "GET":
            continue

        if URL.startswith(STATIC_PREFIXES) or URL.endswith(STATIC_EXTS):
            continue

        out.write(line + "\n")
```
